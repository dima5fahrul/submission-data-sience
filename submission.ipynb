### Cleaning Data

#### a. Changping

```python
columns_to_interpolate = ["PM2.5", "PM10", "SO2", "NO2", "CO", "O3", "TEMP", "PRES", "DEWP", "RAIN", "WSPM"]

for col in columns_to_interpolate:
    changping_df[col] = changping_df[col].interpolate(method="linear", limit_direction="forward")

changping_df.isna().sum()
```

```python
wind_direction_map = {
    'N': 0, 'NNE': 22.5, 'NE': 45, 'ENE': 67.5,
    'E': 90, 'ESE': 112.5, 'SE': 135, 'SSE': 157.5,
    'S': 180, 'SSW': 202.5, 'SW': 225, 'WSW': 247.5,
    'W': 270, 'WNW': 292.5, 'NW': 315, 'NNW': 337.5
}

changping_df["wd"] = changping_df["wd"].map(wind_direction_map)

mean_wd_degrees = changping_df["wd"].mean()
print("Mean wind direction in degrees: ", mean_wd_degrees)
```

```python
changping_df["wd"] = changping_df["wd"].interpolate(method="linear", limit_direction="forward")
changping_df.isna().sum()
```

```python
wind_direction_reverse_map = {
    0: 'N', 22.5: 'NNE', 45: 'NE', 67.5: 'ENE',
    90: 'E', 112.5: 'ESE', 135: 'SE', 157.5: 'SSE',
    180: 'S', 202.5: 'SSW', 225: 'SW', 247.5: 'WSW',
    270: 'W', 292.5: 'WNW', 315: 'NW', 337.5: 'NNW'
}

changping_df["wd"] = changping_df["wd"].map(wind_direction_reverse_map)

changping_df[changping_df.wd.isna()]
```

```python
changping_df.wd = changping_df.wd.fillna(value="S")
changping_df.wd.isna().sum()
```

#### b. Dingling

```python
for col in columns_to_interpolate:
    dingling_df[col] = dingling_df[col].interpolate(method="linear", limit_direction="forward")
    
dingling_df.isna().sum()
```

```python
dingling_df["wd"] = dingling_df["wd"].map(wind_direction_map)
dingling_df["wd"] = dingling_df["wd"].interpolate(method="linear", limit_direction="forward")

dingling_mean_wd_degrees = dingling_df["wd"].mean()
print("Mean wind direction in degrees: ", dingling_mean_wd_degrees)
dingling_df.isna().sum()
```

```python
dingling_df[dingling_df.NO2.isna()]
```

```python
dingling_df.NO2 = dingling_df.NO2.fillna(value=dingling_df.NO2.mean())
dingling_df.NO2.isna().sum()
```

```python
dingling_df["wd"] = dingling_df["wd"].map(wind_direction_reverse_map)
dingling_df["wd"] = dingling_df["wd"].fillna(value="S")

dingling_df.wd.isna().sum()
```

#### c. Dongsi

```python
for col in columns_to_interpolate:
    dongsi_df[col] = dongsi_df[col].interpolate(method="linear", limit_direction="forward")
    
dongsi_df.isna().sum()
```

```python
dongsi_df.wd = dongsi_df.wd.map(wind_direction_map)
dongsi_df.wd = dongsi_df.wd.interpolate(method="linear", limit_direction="forward")

dongsi_df.wd.isna().sum()
```

```python
dongsi_df.wd.mean()
```

```python
dongsi_df.wd = dongsi_df.wd.map(wind_direction_reverse_map)
dongsi_df.wd.isna().sum()
```

```python
dongsi_df.wd = dongsi_df.wd.fillna(value="SSE")
dongsi_df.wd.isna().sum()
```

#### 4. Gucheng

```python
for col in columns_to_interpolate:
    gucheng_df[col] = gucheng_df[col].interpolate(method='linear', limit_direction='forward')

gucheng_df.isna().sum()
```

```python
gucheng_df.NO2 = gucheng_df.NO2.fillna(gucheng_df.NO2.mean())
gucheng_df.NO2.isna().sum()
```

```python
gucheng_df.wd = gucheng_df.wd.map(wind_direction_map)
gucheng_df.wd = gucheng_df.wd.interpolate(method="linear", limit_direction="forward")

gucheng_df.wd.isna().sum()
gucheng_df.wd.mean()
```

```python
gucheng_df.wd = gucheng_df.wd.map(wind_direction_reverse_map)
gucheng_df.wd.isna().sum()
```

```python
gucheng_df.wd = gucheng_df.wd.fillna(value="SSE")
gucheng_df.wd.isna().sum()
```

### Pertanyaan 3: Bagaimana perbandingan rata-rata kualitas partikel udara PM2.5 (µg/m³) harian dalam satu minggu di kota Changping, Dingling, Dongsi, dan Gucheng?

#### a. Changping

```python
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Changping
daily_avg_pm25_changping = changping_df['PM2.5'].resample('D').mean()
daily_avg_pm25_changping = daily_avg_pm25_changping.reset_index()

daily_avg_pm25_changping['day_of_week'] = daily_avg_pm25_changping['date'].dt.day_name()
daily_avg_pm25_changping['day_of_week'] = pd.Categorical(daily_avg_pm25_changping['day_of_week'], categories=days_order, ordered=True)

daily_average_pm25_by_day_changping = daily_avg_pm25_changping.groupby('day_of_week', observed=False)['PM2.5'].mean()

plt.figure(figsize=(10, 6))
daily_average_pm25_by_day_changping.plot(kind='bar', color='skyblue')
plt.title('PM2.5 (µg/m³) Daily Average by Day of Week in Changping', fontsize=14)
plt.xlabel('Day of Week', fontsize=12)
plt.ylabel('PM2.5 (µg/m³)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#### b. Dingling

```python
# Dingling
daily_avg_pm25_dingling = dingling_df['PM2.5'].resample('D').mean()
daily_avg_pm25_dingling = daily_avg_pm25_dingling.reset_index()

daily_avg_pm25_dingling['day_of_week'] = daily_avg_pm25_dingling['date'].dt.day_name()
daily_avg_pm25_dingling['day_of_week'] = pd.Categorical(daily_avg_pm25_dingling['day_of_week'], categories=days_order, ordered=True)

daily_average_pm25_by_day_dingling = daily_avg_pm25_dingling.groupby('day_of_week', observed=False)['PM2.5'].mean()

plt.figure(figsize=(10, 6))
daily_average_pm25_by_day_dingling.plot(kind='bar', color='lightcoral')
plt.title('PM2.5 (µg/m³) Daily Average by Day of Week in Dingli', fontsize=14)
plt.xlabel('Day of Week', fontsize=12)
plt.ylabel('PM2.5 (µg/m³)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
